```python
#Feature Engineering
plt.figure(figsize=(12, 6))
corr = imp[num_cols + ['got_clk', 'got_conv']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
from sklearn.feature_selection import mutual_info_classif

# Select One-Hot Encoded Columns
ohe_cols = imp.columns[imp.columns.str.startswith(('Long-term', 'In-market', 'Demographic'))]

# Compute Mutual Information
mi = mutual_info_classif(imp[ohe_cols], imp['got_conv'])

# Convert to Pandas Series for Sorting & Visualization
mi_series = pd.Series(mi, index=ohe_cols).sort_values(ascending=False).head(10)

# Plot the Top 10 Features by Mutual Information
mi_series.plot(kind='barh', figsize=(12, 6), color='teal')
plt.title("Top 10 Most Important One-Hot Encoded Features (Mutual Info)")
plt.xlabel("Mutual Information Score")
plt.show()
```

```python
#dimensionality reduction and feature selection

#PRINCIPAL COMPONENT ANALYSIS
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
ohe_scaled = scaler.fit_transform(imp[ohe_cols])

pca = PCA(n_components=3)
pca_result = pca.fit_transform(ohe_scaled)

plt.figure(figsize=(10, 5))
plt.scatter(pca_result[:, 0], pca_result[:, 1], alpha=0.3, c=imp['got_conv'], cmap='coolwarm')
plt.colorbar(label='Conversion')
plt.title("PCA: Visualizing One-Hot Encoded Features in 2D")
plt.show()
```

```python
from sklearn.feature_selection import VarianceThreshold
   
selector = VarianceThreshold(threshold=0.01)  # Drop features with low variance
selector.fit(imp[ohe_cols])
low_var_cols = ohe_cols[~selector.get_support()]
print("Low Variance Features to Drop:", low_var_cols.tolist())
```

```python
upper_tri = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.85)]
print("Highly Correlated Features to Drop:", to_drop)
```
