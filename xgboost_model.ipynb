```python
import pandas as pd

column_names_bid = [
    "BidID", "Timestamp", "VisitorID", "User-Agent", "IP", "Region", "City",
    "Adexchange", "Domain", "URL", "AnonymousURLID", "AdslotID", "Adslotwidth",
    "Adslotheight", "Adslotvisibility", "Adslotformat", "Adslotfloorprice",
    "CreativeID", "Biddingprice", "AdvertiserID", "User_tag"
]

column_names = [
    "BidID", "Timestamp", "Logtype", "VisitorID", "User-Agent", "IP", "Region", "City",
    "Adexchange", "Domain", "URL", "AnonymousURLID", "AdslotID", "Adslotwidth",
    "Adslotheight", "Adslotvisibility", "Adslotformat", "Adslotfloorprice",
    "CreativeID", "Biddingprice", "Payingprice", "KeypageURL", "AdvertiserID", "User_tag"
]


bid = pd.read_csv("dataset/bid.07.txt", delimiter='\t', names=column_names_bid,low_memory=True)
clk = pd.read_csv("dataset/clk.07.txt", delimiter='\t',names=column_names ,low_memory=True)
conv = pd.read_csv("dataset/conv.07.txt", delimiter='\t',names=column_names, low_memory=True)
imp = pd.read_csv("dataset/imp.07.txt", delimiter='\t',names=column_names, low_memory=True)
```

```python
print(f"Bid txt shape         : {bid.shape}")
print(f"Imporession txt shape : {imp.shape}")
print(f"Click txt shape       : {clk.shape}")
print(f"Conversion txt shape  : {conv.shape}")
```

```python
all_dfs = [bid,clk,conv,imp]
strings = ["bid","clk","conv","imp"]
for i,df in enumerate(all_dfs):
    for column in df.columns:
        if df[column].isnull().sum() == df[column].shape[0]:
            print(strings[i] , column)
```

```python
bid['has_impression'] = bid['BidID'].isin(imp['BidID'])
```

```python
imp['got_clk'] = imp['BidID'].isin(clk['BidID'])
print(imp.loc[imp['got_clk'] == True].shape)
imp.loc[imp['got_clk'] == True].head()
```

```python
imp['got_conv'] = imp['BidID'].isin(conv['BidID'])
print(imp.loc[imp['got_conv'] == True].shape)
imp.loc[imp['got_conv'] == True].head()
```

```python
imp = imp.drop(['Logtype', 'VisitorID', 'User-Agent', 'IP', 'Adexchange', 'Domain', 'URL', 'AnonymousURLID', 'Adslotfloorprice', 'Biddingprice', 'KeypageURL', 'Timestamp'], axis=1)
imp.head()
```

```python
with open('Adobe Devcraft PS/user.profile.tags.txt') as f:
    tag_dict = {}
    for idx, line in enumerate(f.readlines()):
        tag_dict[line[:5]] = line[6:]

print(tag_dict)
```

```python
imp['User_tag'] = imp['User_tag'].str.split(',')
```

```python
imp = imp.explode('User_tag')
imp = pd.get_dummies(imp, columns=['User_tag'])
imp = imp.groupby('BidID', as_index=False).max()
```

```python
column_names = list(imp.columns)
column_names
```

```python
for i, col in enumerate(column_names):
    if(not col[:4] == 'User'):
        continue
    
    column_names[i] = tag_dict[col[-5:]][:-1]
column_names
```

```python
imp.columns = column_names
```

```python
for key in tag_dict:
    if(tag_dict[key][:-1] not in column_names):
        print(tag_dict[key][:-1])
        imp[tag_dict[key][:-1]] = False
```

```python
from sklearn.preprocessing import LabelEncoder

imp = imp.drop(columns=['BidID', 'Payingprice'], axis=1)

categorical_features = ['Region', 'City', 'Adslotvisibility', 'Adslotformat', 'CreativeID', 'AdvertiserID', 'AdslotID']

for col in categorical_features:
    le = LabelEncoder()
    imp[col] = le.fit_transform(imp[col])

boolean_features = [col for col in imp.columns if imp[col].dtype == 'bool']
imp[boolean_features] = imp[boolean_features].astype(int)
```

```python
imp.head()
```

```python
from imblearn.over_sampling import SMOTE

y = imp['got_conv']
X = imp.drop(columns=['got_conv'])

smote = SMOTE(sampling_strategy=0.15)
X, y = smote.fit_resample(X, y)

imp_resampled = pd.DataFrame(X, columns=X.columns)
imp_resampled['got_conv'] = y

imp_resampled.sample(10)
```

```python
from sklearn.utils import resample

clk_class = imp[imp['got_clk'] == 1]

got_clk_resample= resample(clk_class, replace=True, n_samples=700000)

imp_oversampled = pd.concat([imp_resampled, got_clk_resample])
```

```python
print("Shape:", imp_oversampled.shape)
print("Number of clicked impressions:",imp_oversampled['got_clk'].sum())
print("Number of converted impressions:",imp_oversampled['got_conv'].sum())
```

```python
X = imp_oversampled.drop(['got_conv', 'got_clk'], axis=1)
clk_label = imp_oversampled['got_clk']
conv_label = imp_oversampled['got_conv']
```

```python
# Train for got_clk
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

X_train, X_test, y_train, y_test = train_test_split(X, clk_label, test_size=0.1, stratify=clk_label)
clk_model = xgb.XGBClassifier(objective='binary:logistic', eval_metric='auc', use_label_encoder=False, n_estimators=60, max_depth=6)
clk_model.fit(X_train, y_train)

# Evaluate
clk_pred_prob = clk_model.predict_proba(X_test)[:, 1]  # Get probability of got_clk = 1
auc_score = roc_auc_score(y_test, clk_pred_prob)
print(f"Click Model AUC: {auc_score}")
```

```python
from sklearn.metrics import confusion_matrix

clk_pred = clk_model.predict(X_test)
cm1 = confusion_matrix(y_test, clk_pred)
print(cm1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, conv_label, test_size=0.1, stratify=conv_label)
conv_model = xgb.XGBClassifier(objective='binary:logistic', eval_metric='auc', use_label_encoder=False, n_estimators=60, max_depth=6)
conv_model.fit(X_train, y_train)

# Evaluate
conv_pred_prob = conv_model.predict_proba(X_test)[:, 1]  # Get probability of got_conv = 1
auc_score = roc_auc_score(y_test, clk_pred_prob)
print(f"Click Model AUC: {auc_score}")

conv_pred = conv_model.predict(X_test)
cm2 = confusion_matrix(y_test, conv_pred)
print(cm2)
```
